```python
import matplotlib.pyplot as plt
country_count=admitted['School Country'].value_counts()[:20]
city_count=admitted['School City'].value_counts()[:20]
```

```python
# Create the bar chart
country_count.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Value Counts')
plt.show()
```

```python
# Create the bar chart
city_count.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Value Counts')
plt.show()
```

```python
us_city=admitted[admitted['School Country']=='United States']
us_city_count=us_city['School City'].value_counts()[:20]
```

```python
# Create the bar chart
us_city_count.plot(kind='bar')
plt.xlabel('U.S. Cities')
plt.ylabel('Count of Admitted Student')
plt.title('Transfer Students HS location: Top 20 U.S. Cities')
plt.show()
```

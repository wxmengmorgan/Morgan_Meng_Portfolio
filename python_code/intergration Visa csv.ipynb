```python
import os
import csv

# Directories
input_folder = '/Users/morgan/Downloads/visa by month/visa by month CSV/Monthly Visa Stats by Nationality'  # Replace with the path to your input folder
output_folder = 'Users/morgan/Downloads/visa by month/visa by month CSV/Processed'  # Replace with the desired output folder
os.makedirs(output_folder, exist_ok=True)  # Create output folder if it doesn't exist
```

```python
# Function to check if the file is already processed (3 columns)
def is_already_processed(csv_file_path):
    try:
        with open(csv_file_path, 'r', newline='', encoding='utf-8') as file:
            reader = csv.reader(file)
            for row in reader:
                if len(row) != 3:  # If any row doesn't have exactly 3 columns, not processed
                    return False
            return True  # If all rows have exactly 3 columns, it's processed
    except Exception as e:
        print(f"Error checking file {csv_file_path}: {e}")
        return False  # If there's an issue reading the file, assume not processed
```

```python
def remove_repeated_headers(rows):
    """
    Remove repeated headers and duplicate rows based on patterns.
    Keep the first occurrence of each detected header.
    """
    clean_rows = []
    seen_rows = set()  # To track already processed rows
    header_keywords = ["Nonimmigrant", "Nationality", "Visa","("]  # Add more keywords if needed

    for i, row in enumerate(rows):
        # Convert row to a string to check for header keywords and duplicate rows
        row_content = " ".join(cell.strip() for cell in row)

        # Skip duplicates but keep the first occurrence
        if row_content in seen_rows:
            print(f"Skipping duplicate row: {row_content}")
            continue
        else:
            seen_rows.add(row_content)

        # Skip rows matching header patterns
        if any(keyword in row_content for keyword in header_keywords):
            print(f"Skipping detected header: Row {i} -> {row_content}")
            continue

        # Otherwise, add the row to the clean list
        clean_rows.append(row)
    
    return clean_rows
```

```python
# Function to split a string by first and second spaces from the back
def split_row_by_spaces(row):
    try:
        # Split by spaces from the back
        parts = row.rsplit(' ', 2)  # Splits into [start_of_row, second_to_last, last]
        if len(parts) == 3:
            return parts  # Return the three parts as a list
        else:
            return [row]  # Return the original row if it cannot be split
    except Exception as e:
        print(f"Error splitting row: {e}")
        return [row]  # Return the original row on error
```

```python
# Process each CSV file in the input folder
for csv_file in os.listdir(input_folder):
    if csv_file.endswith('.csv'):
        input_path = os.path.join(input_folder, csv_file)
        output_path = os.path.join(output_folder, csv_file)

        # Skip if already processed
        if is_already_processed(input_path):
            print(f"Skipping already processed file: {csv_file}")
            continue

        try:
            print(f"Processing {csv_file}...")
            
            # Read the CSV file
            with open(input_path, 'r', newline='', encoding='utf-8') as infile:
                reader = list(csv.reader(infile))
                print(f"Original rows in {csv_file}: {len(reader)}")
                
                # Remove repeated headers
                cleaned_rows = remove_repeated_headers(reader)
                print(f"Rows after removing headers: {len(cleaned_rows)}")
                
                # Transform rows (split each row starting from the first)
                transformed_rows = []
                for row in cleaned_rows:
                    joined_row = ' '.join(row)  # Rejoin row for splitting
                    transformed_rows.append(split_row_by_spaces(joined_row))
            
            # Write the transformed data to a new CSV file
            with open(output_path, 'w', newline='', encoding='utf-8') as outfile:
                writer = csv.writer(outfile)
                writer.writerows(transformed_rows)  # Write the transformed rows

            print(f"Saved transformed CSV to {output_path}")
        
        except Exception as e:
            print(f"Error processing {csv_file}: {e}")
```

```python
# Function to detect and remove repeated headers
def detect_and_remove_repeated_headers(csv_file_path):
    """
    Detects and removes repeated headers in a CSV file.
    Returns a cleaned list of rows.
    """
    try:
        with open(csv_file_path, 'r', newline='', encoding='utf-8') as file:
            reader = list(csv.reader(file))
            
        clean_rows = []
        seen_headers = set()
        header_keywords = ["Nonimmigrant", "Nationality", "Visa"]  # Patterns to detect headers
        
        for i, row in enumerate(reader):
            row_content = " ".join(cell.strip() for cell in row)  # Combine cells into one line
            
            # Detect if the row matches a header pattern
            is_header = any(keyword in row_content for keyword in header_keywords)
            
            if is_header:
                if row_content in seen_headers:
                    print(f"Removing repeated header: {row_content}")
                    continue  # Skip repeated headers
                else:
                    seen_headers.add(row_content)  # Keep the first occurrence
            
            clean_rows.append(row)  # Add the row to the cleaned list if not skipped

        print(f"Cleaned rows: {len(clean_rows)} / Original rows: {len(reader)}")
        return clean_rows

    except Exception as e:
        print(f"Error processing {csv_file_path}: {e}")
        return None

# Write cleaned rows back to the same file (or a new one)
def write_cleaned_csv(csv_file_path, cleaned_rows):
    """
    Writes cleaned rows back to a CSV file.
    """
    try:
        with open(csv_file_path, 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerows(cleaned_rows)
        print(f"Cleaned CSV saved: {csv_file_path}")
    except Exception as e:
        print(f"Error saving cleaned file: {e}")

# Process all CSV files in a folder
def process_folder_for_headers(folder_path):
    """
    Detects and removes repeated headers for all CSV files in a folder.
    """
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):  # Only process CSV files
            file_path = os.path.join(folder_path, file_name)
            print(f"Processing {file_name}...")
            cleaned_rows = detect_and_remove_repeated_headers(file_path)  # Clean headers
            if cleaned_rows:
                write_cleaned_csv(file_path, cleaned_rows)  # Save the cleaned file
```

```python
# Set the folder path and process all CSV files
folder_path = '/Users/morgan/Downloads/visa by month/visa by month CSV/Monthly Visa Stats by Post/2024'  # Replace with your folder path
process_folder_for_headers(folder_path)
```

```python
folder_path = '/Users/morgan/Downloads/visa by month/visa by month CSV/Monthly Visa Stats by Post/2024' 
# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Only process CSV files
        # Check if '%20' exists in the filename
        if '%20' in file_name:
            # Replace '%20' with a space
            new_file_name = file_name.replace('%20', ' ')
            # Construct full paths
            old_file_path = os.path.join(folder_path, file_name)
            new_file_path = os.path.join(folder_path, new_file_name)
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: "{file_name}" -> "{new_file_name}"')
        else:
            print(f'No changes for: "{file_name}"')
```
